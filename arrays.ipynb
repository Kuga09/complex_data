```python
import matplotlib.pyplot as plt
x=list(range(-10,11))
y=[]
for i in x:
  y.append(i**2) 
plt.title('Парабола') 
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.plot(x,y,"-.", marker='o', markersize=3)
plt.show()
```
